```python
def assign_season(date_str):
    try:
        # Attempting to parse the date in 'dd/mm/yyyy' format
        date_obj = pd.to_datetime(date_str, format='%d/%m/%Y')
    except ValueError:
        try:
            # If that fails, attempting to parse the date in 'dd/mm/yy' format
            date_obj = pd.to_datetime(date_str, format='%d/%m/%y')
        except ValueError:
            return 'Unknown'

    for season, season_dates in seasons.items():
        start_date = pd.to_datetime(season_dates['start_date'], format='%d-%m-%Y')
        end_date = pd.to_datetime(season_dates['end_date'], format='%d-%m-%Y')
        if start_date <= date_obj <= end_date:
            return season
    return 'Unknown'

# Applying the function to assign the season to each date in the 'Date' column 
selected_data['Season'] = selected_data['Date'].apply(assign_season)

selected_data.info()
```

```python
season_data = {}

for season in seasons:
   
    season_data = selected_data[selected_data['Season'] == season].copy()

    
    season_data['Points'] = season_data.apply(lambda row: 3 if row['FTR'] == 'H' else 1 if row['FTR'] == 'D' else 0, axis=1)
    season_data['GoalsScored'] = season_data['FTHG']
    season_data['GoalsConceded'] = season_data['FTAG']
    season_data['GoalDifference'] = season_data['FTHG'] - season_data['FTAG']
    
    team_points = {}
    team_wins = {}
    team_losses = {}
    team_draws = {}
    team_goals_scored = {}
    team_goals_conceded = {}
    team_goal_difference = {}
    team_home_shots = {}
    team_away_shots={}
    team_home_wins={}
    team_away_wins={}
    team_home_shots_on_target={}
    team_away_shots_on_target={}
    team_home_yellow_cards={}
    team_away_yellow_cards={}
    team_home_red_cards={}
    team_away_red_cards={}
    

    
    for index, row in season_data.iterrows():
        home_team = row['HomeTeam']
        away_team = row['AwayTeam']

        if home_team not in team_points: 
            team_points[home_team] = 0
            team_wins[home_team] = 0
            team_losses[home_team] = 0
            team_draws[home_team] = 0
            team_goals_scored[home_team] = 0
            team_goals_conceded[home_team] = 0
            team_goal_difference[home_team] = 0
            team_home_shots[home_team]= 0
            team_away_shots[home_team]=0
            team_home_wins[home_team] = 0
            team_away_wins[home_team] = 0
            team_home_shots_on_target[home_team]=0
            team_away_shots_on_target[home_team]=0
            team_home_yellow_cards[home_team]=0
            team_away_yellow_cards[home_team]=0
            team_home_red_cards[home_team]=0
            team_away_red_cards[home_team]=0
# The case when team some of the teams are not participating since
# 42 teams participate across 20 seasons while only 20 take part in a season           

        if away_team not in team_points:
            team_points[away_team] = 0
            team_wins[away_team] = 0
            team_losses[away_team] = 0
            team_draws[away_team] = 0
            team_goals_scored[away_team] = 0
            team_goals_conceded[away_team] = 0
            team_goal_difference[away_team] = 0
            team_home_shots[away_team]=0
            team_away_shots[away_team]=0
            team_home_wins[away_team] = 0
            team_away_wins[away_team] = 0
            team_home_shots_on_target[away_team]=0
            team_away_shots_on_target[away_team]=0
            team_home_yellow_cards[away_team]=0
            team_away_yellow_cards[away_team]=0
            team_home_red_cards[away_team]=0
            team_away_red_cards[away_team]=0
            
        if row['FTR'] == 'H':
            team_points[home_team] += 3
            team_wins[home_team] += 1
            team_losses[away_team] += 1
            team_home_wins[home_team] += 1  
        elif row['FTR'] == 'A':
            team_points[away_team] += 3
            team_wins[away_team] += 1
            team_losses[home_team] += 1
            team_away_wins[away_team] += 1  
        else:
            team_points[home_team] += 1
            team_points[away_team] += 1
            team_draws[home_team] += 1
            team_draws[away_team] += 1

        # Updating goals scored, goals conceded, and goal difference
        team_goals_scored[home_team] += row['FTHG']
        team_goals_conceded[home_team] += row['FTAG']
        team_goal_difference[home_team] += row['FTHG'] - row['FTAG']
       
        team_goals_scored[away_team] += row['FTAG']
        team_goals_conceded[away_team] += row['FTHG']
        team_goal_difference[away_team] += row['FTAG'] - row['FTHG']
        
        # Updating total shots and total shots on target
        team_home_shots[home_team] += row['HS']
        team_home_shots[away_team] += row['HS']
        team_away_shots[away_team] += row['AS']
        team_away_shots[home_team] += row['AS']
        team_home_shots_on_target[home_team] += row['HST']
        team_home_shots_on_target[away_team] += row['HST']
        team_away_shots_on_target[home_team] += row['AST']
        team_away_shots_on_target[away_team] += row['AST']
        team_home_yellow_cards[home_team] += row['HY']
        team_home_yellow_cards[away_team] += row['HY']
        team_away_yellow_cards[away_team] += row['AY']
        team_away_yellow_cards[home_team] += row['AY']
        team_home_red_cards[home_team] += row['HR']
        team_home_red_cards[away_team] += row['HR']
        team_away_red_cards[away_team] += row['AR']
        team_away_red_cards[home_team] += row['AR']
        
        
    # Converting the team points dictionary to a DataFrame
    points_df = pd.DataFrame({'Team': list(team_points.keys()), 'Points': list(team_points.values())})

    season_year = int(season_data['Season'].iloc[0].split('-')[0])
    
    # Sorting the points DataFrame by Points in descending order
    points_df = points_df.sort_values(by='Points', ascending=False)

    points_df.reset_index(drop=True, inplace=True)
    
    points_df['Rank'] = points_df.index + 1 #Adding rank column to points_df 


    # Adding the 'Year' column to the points DataFrame
    points_df['Year'] = season_year
    # Adding columns for wins (W), losses (L), draws (D),goals scored,goals conceded,goal difference and other stats
    points_df['W'] = [team_wins[team] for team in points_df['Team']]
    points_df['L'] = [team_losses[team] for team in points_df['Team']]
    points_df['D'] = [team_draws[team] for team in points_df['Team']]
    points_df['GS'] = [team_goals_scored[team] for team in points_df['Team']]
    points_df['GC'] = [team_goals_conceded[team] for team in points_df['Team']]
    points_df['GD'] = [team_goal_difference[team] for team in points_df['Team']]
    points_df['HomeShots'] = [team_home_shots[team] for team in points_df['Team']]
    points_df['Homeshotsontarget'] = [team_home_shots_on_target[team] for team in points_df['Team']]
    points_df['HomeWins'] = [team_home_wins[team] for team in points_df['Team']]
    points_df['AwayWins'] = [team_away_wins[team] for team in points_df['Team']]
    points_df['AwayShots'] = [team_away_shots[team] for team in points_df['Team']]
    points_df['Awayshotsontarget'] = [team_away_shots_on_target[team] for team in points_df['Team']]
    points_df['totalshots'] = points_df['HomeShots'] +points_df['AwayShots']
    points_df['Shotsontarget']= points_df['Homeshotsontarget'] + points_df['Awayshotsontarget']
    points_df['awayyellow']= [team_away_yellow_cards[team] for team in points_df['Team']]
    points_df['homeyellow']= [team_home_yellow_cards[team] for team in points_df['Team']]
    points_df['awayred']= [team_away_red_cards[team] for team in points_df['Team']]
    points_df['homered']= [team_home_red_cards[team] for team in points_df['Team']]
    
 
    
    # Appending the points DataFrame to the list
    season_points_dfs.append(points_df)
    


season_points_dfs[1]
```

```python
import matplotlib.pyplot as plt
import pandas as pd

team1_name = "Man City"
team2_name = "Man United"

years = []
team1_goal_efficiency = []
team2_goal_efficiency = []

for season_points_df in season_points_dfs:
    # Filter the DataFrame for the first team (Man City)
    team1_data = season_points_df[season_points_df['Team'] == team1_name]
    
    # Filter the DataFrame for the second team (Man United)
    team2_data = season_points_df[season_points_df['Team'] == team2_name]
    
    # Checking if both teams exist in the current season
    if not team1_data.empty and not team2_data.empty:
        # Calculating goal scoring efficiency for the first team (Man City)
        goals_scored_team1 = team1_data.iloc[0]['GS']
        shots_on_target_team1 = team1_data.iloc[0]['Shotsontarget']
        
        if shots_on_target_team1 > 0:
            efficiency_team1 = goals_scored_team1 / shots_on_target_team1
        else:
            efficiency_team1 = 0

        # Calculating goal scoring efficiency for the second team (Man United)
        goals_scored_team2 = team2_data.iloc[0]['GS']
        shots_on_target_team2 = team2_data.iloc[0]['Shotsontarget']
        
        if shots_on_target_team2 > 0:
            efficiency_team2 = goals_scored_team2 / shots_on_target_team2
        else:
            efficiency_team2 = 0
        
        # Getting the year for the current season
        year = team1_data.iloc[0]['Year']
        
        # Adding the year and goal scoring efficiency for both teams to the lists
        years.append(year)
        team1_goal_efficiency.append(efficiency_team1)
        team2_goal_efficiency.append(efficiency_team2)

# Sorting the years and goal_efficiency lists by year
years, team1_goal_efficiency, team2_goal_efficiency = zip(*sorted(zip(years, team1_goal_efficiency, team2_goal_efficiency)))

# Create a line plot to visualize goal scoring efficiency for both teams
plt.figure(figsize=(10, 6))
plt.plot(years, team1_goal_efficiency, marker='o', linestyle='-', label=team1_name)
plt.plot(years, team2_goal_efficiency, marker='o', linestyle='-', label=team2_name)
plt.title(f"{team1_name} vs. {team2_name} - Goal Scoring Efficiency Across Seasons")
plt.xlabel("Year")
plt.ylabel("Goal Scoring Efficiency (Goals Scored per Shot on Target)")
plt.grid(True)

# Set the x-axis to display integer values without decimal places
plt.xticks(range(int(min(years)), int(max(years) + 1), 1))

# Add a legend to distinguish between the two teams
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

all_seasons_points = pd.concat(season_points_dfs)

# Calculating the most successful teams based on consistent top 4 finishes
top_teams = all_seasons_points.groupby('Team').apply(lambda x: (x['Rank'] <= 4).sum()).reset_index(name='Top4Count')
consistent_top_teams = top_teams[top_teams['Top4Count'] >= 10]

plt.figure(figsize=(12, 6))
sns.barplot(x=consistent_top_teams['Team'], y='Top4Count', data=consistent_top_teams, palette='viridis')
plt.title("The Most Consistent Teams")
plt.xlabel("Team")
plt.ylabel("Number of Top 4 Finishes")
plt.xticks(rotation=45)
plt.show()
```

```python
bottom_teams = all_seasons_points.groupby('Team').apply(lambda x: (x['Rank'] > len(x) - 4).sum()).reset_index(name='Bottom4Count')
consistent_bottom_teams = bottom_teams[bottom_teams['Bottom4Count'] >= 10] 

plt.figure(figsize=(12, 6))
sns.barplot(x=consistent_bottom_teams['Team'], y='Bottom4Count', data=consistent_bottom_teams, palette='viridis')
plt.title("The Least Consistent Teams (Bottom 4 Finishes)")
plt.xlabel("Team")
plt.ylabel("Number of Bottom 4 Finishes")
plt.xticks(rotation=45)
plt.show()
```
